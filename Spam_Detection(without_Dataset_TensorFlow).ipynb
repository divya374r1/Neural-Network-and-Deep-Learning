```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np

# Sample email text data
emails = [
    "Get free tickets now!",
    "Limited offer, buy now!",
    "Your wing is waiting",
    "Not spam",
    "Free trial, sign up now!"
]

# Labels: 1 = spam, 0 = not spam
labels = [1, 1, 1, 0, 1]

# Tokenizing the text
tokenizer = Tokenizer()                 # Create tokenizer
tokenizer.fit_on_texts(emails)          # Learn the words
sequences = tokenizer.texts_to_sequences(emails)  # Convert words to numbers

# Padding sequences
max_length = 10                          # Fixed length for all emails
padded_sequences = pad_sequences(sequences, maxlen=max_length, padding='post')

# Building the model
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(input_dim=100, output_dim=10, input_length=max_length),  # Word embeddings
    tf.keras.layers.Flatten(),                    # Convert 2D to 1D
    tf.keras.layers.Dense(64, activation='relu'), # Hidden layer
    tf.keras.layers.Dense(1, activation='sigmoid')# Output: spam or not
])

# Compiling the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Training the model
model.fit(padded_sequences, np.array(labels), epochs=10)

# Testing the model
test_email = ["Get free stuff now!"]                 # New email
test_sequence = tokenizer.texts_to_sequences(test_email)  # Convert to numbers
test_padded_sequence = pad_sequences(test_sequence, maxlen=max_length, padding='post')
prediction = model.predict(test_padded_sequence)     # Predict spam or not
# Print result
print("Spam" if prediction > 0.5 else "Not Spam")
```
